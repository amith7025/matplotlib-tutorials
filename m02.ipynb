```python
import matplotlib.pyplot as plt

# Create a figure and subplots
# The arguments define the number of rows, number of columns, and the subplot index
# Indexing starts from 1 and increases left to right, top to bottom
fig = plt.figure()

# Adding the first subplot
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([1, 2, 3], [4, 5, 6], label='Line 1')
ax1.set_title('Subplot 1')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()

# Adding the second subplot
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter([3, 4, 5], [2, 1, 3], color='red', label='Points')
ax2.set_title('Subplot 2')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.legend()

# Adding the third subplot
ax3 = fig.add_subplot(2, 2, 3)
ax3.bar(['A', 'B', 'C'], [8, 6, 10], color='green', label='Bars')
ax3.set_title('Subplot 3')
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Y-axis')
ax3.legend()

# Adjust layout to prevent overlapping of subplots
plt.tight_layout()

# Display the figure
plt.show()
```
