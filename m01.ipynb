```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

basic graph

```python
x = [1,2,3]
y = [2,4,6]
plt.figure(figsize=(10,7),dpi=300)
plt.plot(x,y,label='linear regression',color='green',linewidth=2,marker='.',markersize=10,markeredgecolor='blue',linestyle='--')

L = np.arange(0,7,0.5)
r = np.sin(L)
plt.plot(L[:4],r[:4],linestyle='--')
plt.plot(L[4:],r[4:],linestyle=None)

plt.title('first graph',fontdict={'fontsize':20})
plt.xlabel('x-axis',fontdict={'fontsize':20})
plt.ylabel('y-axis',fontdict={'fontsize':20})
plt.xticks([1,2,3,4,5,6])
plt.yticks([2,4,5,6,7,9])
plt.legend()
plt.show()
```
